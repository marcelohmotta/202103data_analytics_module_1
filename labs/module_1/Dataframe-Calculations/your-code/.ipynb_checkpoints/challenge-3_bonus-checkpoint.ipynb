## Q1: How to identify VIP & Preferred Customers?

We start by importing all the required libraries:

```python
# import required libraries
import numpy as np
import pandas as pd
```

Next, import `Orders` from the provided data folder. Please access the data  with a relative path.
Load `Orders` into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

```python
# your code here

orders = pd.read_csv('data/Orders.csv')
orders.head()
```

---

"Identify VIP and Preferred Customers" is the non-technical goal of your boss. You need to translate that goal into technical languages that data analysts use:

## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here

rank_cust = orders.groupby(["CustomerID"]).sum()

VIP_quantile = np.quantile(rank_cust["amount_spent"], 0.95)
Pref_quantile = np.quantile(rank_cust["amount_spent"], 0.75)

condition = [rank_cust["amount_spent"] < Pref_quantile, (rank_cust["amount_spent"] >= Pref_quantile) & (rank_cust["amount_spent"] < VIP_quantile), (rank_cust["amount_spent"] >=  VIP_quantile)]
cust_label = ["normal", "preferred", "VIP"]

rank_cust["cust_type"] = np.select(condition, cust_label)

rank_cust.head()
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here

country_merge = rank_cust.merge(orders[["CustomerID", "Country"]], left_on = 'CustomerID', right_on = 'CustomerID', how = 'inner')

country_merge.head()
```

```python
print(len(country_merge), len(orders))
```

```python
country_merge_short = country_merge.drop_duplicates()
print(len(country_merge_short), len(rank_cust))

country_merge_short.head()
```

```python
VIPs = country_merge_short[country_merge_short['cust_type'] == 'VIP']
VIPs.head()
len(VIPs)
```

```python
nr_VIPs = VIPs.groupby(['Country']).count().cust_type
nr_VIPs
nr_VIPs.sort_values(ascending=False)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here

VIP_pref = country_merge_short[country_merge_short['cust_type'] != 'normal']
VIP_pref.head()
```

```python
nr_VIP_pref = VIP_pref.groupby(["Country"]).count().cust_type
nr_VIP_pref
nr_VIP_pref.sort_values(ascending = False)
```
